```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate the denoise factor
def denoise_function(steps, start_at_step):
    return (steps - start_at_step) / steps

# Generating values for plotting
steps = np.arange(1, 21)  # Assuming a range of steps from 1 to 100
start_at_step_values = [1, 5, 10, 15,20]  # Different start_at_step values for demonstration

# Plotting
plt.figure(figsize=(10, 6))
for start_at_step in start_at_step_values:
    denoise_values = denoise_function(steps, start_at_step)
    plt.plot(steps, denoise_values, label=f"start_at_step = {start_at_step}")

plt.xlabel("Steps")
plt.ylabel("Amount of Noise")
plt.title("Denoise Schedule")
plt.legend()
plt.grid(True)
plt.show()
```
